# How to use vaery to generate sythetic data based on the iris dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from vaery.vaery import VAE
from vaery.vaery.preprocessing import VAEPreprocessor
from vaery.vaery import SimpleBaseModel
from ivis import Ivis

import tensorflow as tf
tf.random.set_seed(1234)
```

## 1. Data Import

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
df = pd.DataFrame(X, columns=iris.feature_names)
df['Species'] = y
```

## 2. Create Variational Autoencoder (VAE) instance and fit to data
Note we use the SimpleBaseModel() here due to the lack in complexity of the dataset.

```python
vae = VAE(model=SimpleBaseModel(input_shape=df.shape[1]), preprocessor=VAEPreprocessor(), batch_size=20)
```

```python
vae.fit(df)
```

## 3. Generate synthetic data and format appropriately

```python
vae.sample(df)
```

```python
vae_output = vae.decoded_samples_
```

```python
vae_output["sepal length (cm)"] = [round(x, 1) for x in vae_output["sepal length (cm)"]]
vae_output["sepal width (cm)"] = [round(x, 1) for x in vae_output["sepal width (cm)"]]
vae_output["petal length (cm)"] = [round(x, 1) for x in vae_output["petal length (cm)"]]
vae_output["petal width (cm)"] = [round(x, 1) for x in vae_output["petal width (cm)"]]
vae_output["Species"] = np.abs(np.round(vae_output["Species"]))
```

```python
vae_output
```

## 4. Use ivis dimensionality reduction to visualise data 

```python
from ivis import Ivis
```

```python
ivis = Ivis(k=50)
```

```python
ivis.fit(df.values)
```

```python
embeddings = ivis.transform(df.values)
```

```python
embeddings_synth = ivis.transform(vae_output.values)
```

```python
plt.scatter(embeddings[:,0], embeddings[:,1], label='original')
plt.scatter(embeddings_synth[:,0], embeddings_synth[:,1], label='synthetic')
plt.legend()
plt.xlabel('ivis 1')
plt.ylabel('ivis 2')
```

From this plot we can observe that the sythetic data is comparable to the original data, if somewhat less widely distributed.
